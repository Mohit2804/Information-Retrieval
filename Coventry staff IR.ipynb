```python
import urllib.request 
from bs4 import BeautifulSoup
import requests
import re
from urllib.request import urlopen
import pandas as pd
```

```python
source = requests.get('https://scholar.google.co.uk/citations?view_op=view_org&hl=en&org=9117984065169182779').text
```

```python
soup = BeautifulSoup(source,'lxml')
```

```python
staff = soup.find_all('h3', class_="gs_ai_name")
```

```python
staff
```

```python
publication = soup.find_all('div', class_='gs_ai_int')
```

```python
publication 
```

```python
faculty_name = []
faculty_link = []
for name in staff:
    faculty_name.append(name.text)
    faculty_link.append(name.a.get('href'))
```

```python
faculty_publication = []
for pub in publication:
    faculty_publication.append(pub.text)
```

```python
faculty_name
```

```python
faculty_link
```

```python
faculty_publication
```

```python
d = {'University Name':'Coventry University','faculty_name':faculty_name,'Link':'https://scholar.google.co.uk/','faculty_link':faculty_link,'faculty_publication':faculty_publication}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df["Faculty_URL"] = df["Link"] + df["faculty_link"]
```

```python
df["Faculty_URL"]
```

```python
df2 = df["Link"] + df["faculty_link"]
```

```python
df_final = pd.concat([df, df2], axis=1)
```

```python
df_final
```

```python
df_final.drop(['Link', 'faculty_link'], axis=1, inplace=True)
```
